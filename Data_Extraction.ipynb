# **Assignment Questions**

**Question 1 : Describe different types of data sources used in ETL with suitable examples.**

**Answer:**

ETL processes gather data from diverse sources to consolidate them into a central system. These sources generally fall into the following categories:

* **Relational Databases (RDBMS):** Structured data stored in tables (e.g., MySQL, PostgreSQL, or Oracle).

  Example: A retail company extracting customer transaction records from a SQL database.

* **Flat Files:** Plain text or binary files that contain data in a structured or semi-structured format.

   Example: Daily log files, CSV exports, or Excel spreadsheets.

* **Cloud Applications and APIs:** Data pulled from web services or SaaS platforms.

  Example: Extracting marketing performance data from the Google Ads API.

* **NoSQL Databases:** Unstructured or semi-structured data sources like MongoDB or Cassandra.

  Example: Gathering user social media interactions or sensor data.

**Question 2 : What is data extraction? Explain its role in the ETL pipeline.**

**Answer:**

Data extraction is the first step in the ETL pipeline, involving the retrieval of data from various source systems.


**Its Role**: It acts as the "entry point" for the entire pipeline.


**Purpose:** The primary goal is to collect the necessary data in a way that doesn't negatively impact the performance of the source system, ensuring the data is ready for the transformation phase.

**Question 3 : Explain the difference between CSV and Excel in terms of extraction and ETL usage.**

**solution:**

While both store tabular data, they differ significantly in an ETL context:

* **CSV (Comma Separated Values):** These are plain text files, making them lightweight, fast to parse, and universally compatible. They are preferred for large datasets because they lack formatting overhead.

* **Excel (.xlsx):** These are binary files that support multiple sheets, formulas, and formatting. Extraction is slower because the ETL tool must interpret the file's metadata and structure, which can lead to errors if data types are inconsistent.

**Question 4 : Explain the steps involved in extracting data from a relational database.**

**Answer :**

   Extracting from an RDBMS typically involves these steps:


* **Connection:** Establishing a secure link between the ETL tool and the database using drivers like JDBC or ODBC.

* **Schema Identification:** Identifying the specific tables and columns required for the task.


* **Extraction Logic:** Writing SQL queries (e.g., SELECT statements) to pull the data.

* **Change Data Capture (CDC):** Determining if the extraction should be "Full" (everything) or "Incremental" (only records changed since the last run).

* **Validation**: Checking that the number of records extracted matches the source.

**Question 5 : Explain three common challenges faced during data extraction.**

**Answer:**



* **Source System Performance:** Running heavy extraction queries can slow down the live applications using that database.


* **Data Heterogeneity:** Dealing with inconsistent formats, varying time zones, or different character encodings across multiple sources.


* **Connectivity Issues:** Network instability or credential changes can cause extraction jobs to fail mid-process.

**Question 6 : What are APIs? Explain how APIs help in real-time data extraction.**

**Answer:**

**APIs (Application Programming Interfaces)** allow different software systems to communicate.


**Real-Time Role:** Unlike batch processing (which happens at scheduled intervals), APIs enable streaming extraction.


**Function:** They allow an ETL tool to "subscribe" to data updates or request them the moment a change occurs, facilitating near-instantaneous data availability for dashboards and analytics.

**Question 7 : Why are databases preferred for enterprise-level data extraction?**

**Answer:**

  Enterprises prefer databases over flat files because:

* **Scalability:** Databases can handle terabytes of data with optimized indexing.

* **Security:** They offer robust access controls and encryption.

* **Integrity:** Relational databases enforce data types and relationships, ensuring the extracted data is higher quality.

**Question 8 : What steps should an ETL developer take when extracting data from large CSV files (1GB+)?**

**Answer:**

 When dealing with massive CSVs, an ETL developer should:


* **Use Chunking:** Instead of loading the whole file into memory, read it in smaller "chunks" or blocks.


* **Parallel Processing:** Use multi-threading to process different sections of the file simultaneously.


* **Schema Inference:** Explicitly define data types beforehand to prevent the ETL tool from wasting resources guessing the format.
